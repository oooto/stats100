$$
    f'(x) = \frac{\beta e^{-(\beta_0+x\beta)}}{(1+e^{-(\beta_0+x\beta)})^2} \geq 0
$$
となるため、$f(x)$は単調増加となる。また、
$$
    f''(x) = \frac{-\beta^2e^{-(\beta_0+x\beta)}(1-e^{-(\beta_0+x\beta)})}{(1+e^{-(\beta_0+x\beta)})^3}
$$
となるため、$x<-\beta_0/\beta$で下に凸、$x>-\beta_0/\beta$で上に凸となる。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import japanize_matplotlib
import scipy
from scipy import stats
from numpy.random import randn
```

```python
def f(x):
    return np.exp(beta_0 + beta * x) / (1 + np.exp(beta_0 + beta * x))
```

```python
beta_0 = 0
beta_seq = np.array([0, 0.2, 0.5, 1, 2, 10])
x_seq = np.arange(-10, 10, 0.1)

plt.xlabel("x")
plt.ylabel("P(Y = 1 | x)")
plt.title("ロジスティクス曲線")

for i in range(beta_seq.shape[0]):
    beta = beta_seq[i]
    p = f(x_seq)
    plt.plot(x_seq, p, label="{}".format(beta))
plt.legend(loc="upper left")
```
